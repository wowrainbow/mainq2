# 이상치 찾기

```python
#신용카드번호는 16자리 아닌가?
cc_df['cc_num'].astype(str).str.len().describe()
```

```python
#신용카드번호는 15자리 혹은 16자리여야 한다고 한다.
err_cc = []

for i in cc_df['cc_num']:
    if len(str(i)) not in [15, 16]:
        err_cc.append(i)


#신용카드번호가 맞을 때 T, 아닐때 F 하고 나중에 원핫인코딩 하기
#cc_df = cc_df[~cc_df['cc_num'].isin(err_cc)]]
cc_df['right_cc_num'] = np.where(cc_df['cc_num'].isin(err_cc), 'F', 'T')
```

```python
#거래발생시간 알아보기
cc_time = cc_df[['trans_time', 'unix_time']]
cc_time
```

```python
#대부분 7~8년전이라 무슨 의미인지 알 수 없다. 카드발급일자?
u_to_t = pd.to_datetime(cc_time['unix_time'], unit = 's')
u_to_t
#정확히 7년 차이가 나서 무의미한 컬럼이라고 생각된다.
```

# 피처 엔지니어링

```python
#지리정보 추출
cc_geo = cc_df.iloc[:, [10, 11, 12, 13, 19, 20]].copy()
```

```python
cc_geo.head()
```

```python
#지오파이 라이브러리 설치
# 클라우스 쉘에서 설치함 pip install geopy

from geopy.distance import distance
```

```python
# 거리 계산후에 데이터프레임에 거리 계산 결과 추가
cc_geo['distance_km'] = cc_geo.apply(lambda row: round(distance((row['lat'], row['long']), (row['merch_lat'], row['merch_long'])).km), axis=1)
```

```python
cc_geo.info()
```

```python
#결제당시나이 구하기
#출생일과 결제일을 담은 데이터셋 만들기

cc_df['dob'] = pd.to_datetime(cc_df['dob'])
cc_df['trans_time'] = pd.to_datetime(cc_df['trans_time'])
cc_age = cc_df.iloc[:, [0, 16]].copy()
```

```python
#결제일-출생일 해서 나이 만들기
cc_age['age'] = (cc_age['trans_time'] - cc_age['dob']).dt.days
cc_age['age'] = (cc_age['age'] / 365).astype(int)
```

```python
cc_age.info()
```

# 파이널 데이터셋 만들기

```python
#선택 이유는 회고에 쓰기
sel1 = cc_df.iloc[:, [4, 7, 21, 22]]
sel2 = cc_geo.iloc[:, [6]]
sel3 = cc_age.iloc[:, [2]]   

final_df = pd.merge(sel1, sel2, left_index=True, right_index=True)
final_df = pd.merge(final_df, sel3, left_index=True, right_index=True)
```

```python
final_df
```

```python
#원핫인코딩 준비
final_df = pd.get_dummies(final_df)
```

 시간이 거의 다 지났을 때 샘플 코드를 보게 되었는데 비교해보니 내가 직접 한 것은 초라해 보인다. amt에 z스코어를 적용해 가공할 생각은 해보지 못했다. 거래 시간대별로 나눌 생각은 해보았지만 사기꾼들이 밤낮 가릴 것 같지는 않아서 무시했다. 하지만 곰곰히 생각해보면 의미있는 지표가 될 수 있었을 것이다.
  그리고 distance 계산이 너무 느려서 실망스럽다. 파편화된 데이터셋을 만들어서 좋았던 점은 고치고 싶은 부분을 바로바로 고치고 테스트해볼 수 있다는 점이었다. 중간중간 틀린 부분, 잘못 이해한 부분, 느린 부분을 고칠 수 있었다. 챗gpt의 도움을 받아 distance 계산의 속도를 올려보고 싶었지만 불가능했다.
